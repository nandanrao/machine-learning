# Pricing and hedging under the Binomial model
Consider a Binomial model with $r = 0.04$, $T = 4$, $d = 0.98$, $u = 1.06$, $S0 = 20$, and a European call with maturity $T = 4$ and strike price $K = 20$.

## Arbitrage
The non-arbitrage condition is satisfied in this model and the satisfaction of the condition is given primarily by the fact that for the given values of d, r and u, inequality  
$$
\begin{aligned}
d &< 1+r < u \\
0.98 &< 1.04 < 1.06 \ \text{(true)}
\end{aligned}
$$ 
is satisfied. Moreover, this condition is necessary to obtain a positive risk-neutral probability. Risk-neutrality is condition to have a complete market, which can exist only in absence of arbitrage opportunities. To compute the risk neutral probability we solve the formula 
$$q=\frac{1+r-d}{u-d}=0.75$$
Probability of the price going up is 0.75. Probability that the price will go down is 0.25 ($1 - q$).

## Value
Summing over the conditional probabilities we get the expected value of X. Therefore, we get the following discounted expected value of the option:
$$
\begin{aligned}
V_0&=(1+r)^{-N}\mathbb{E}_p(X) \\
&=\frac{5.25*0.75^4+4(3.34*0.75^3*0.25^1) + 6(1.58*0.75^2*0.25^2)}{1.04^4} \\
&=2.91
\end{aligned}
$$
Such value is equal to the fair fee it should be paid on the market to buy the described call option. This is coherent with what we computed in question *3)*.

# Simulation of a Brownian motion

```r
set.seed(1234, kind = NULL, normal.kind = NULL)

list.of.packages <- c("ggplot2", "latex2exp", "reshape2", "tidyr")
new.packages <- list.of.packages[!(list.of.packages %in% installed.packages()[,"Package"])]
if(length(new.packages)) install.packages(new.packages)

library(ggplot2)
library(latex2exp)
library(reshape2)
library(tidyr)
    
source('~/Development/BGSE/Courses/Statistical Modelling and Inference/repo/easy_ggplot.R')
```

```r
######################################################
#Simulation of Plain Brownian Motion
######################################################
sim.BM <- function(T, n, nsim=1){
    #Inputs I: Time Interval [0,T], Steps in interval n
    #Inputs II: Number of Simulated BMs - nsim
    #Output: Simple Brownian Motion
    time <- seq(0,T,by=(T/n))
    step <- sqrt(T/n)
    realization <- sample(c(-1,1), size=((length(time)-1)*nsim), replace=T, prob=c(0.5,0.5))
    step.realization <- step * realization
    B <- matrix(step.realization, nsim, length(time)-1)
    B <- t(cbind(rep(0, nsim), t(apply(B, 1, cumsum))))
    B <- cbind(time,B)
    data.frame(B)
}

######################################################
#Simulation of Geometric Brownian Motion
######################################################
sim.GBM <- function(mu, sigma, T, n, nsim=1){
    #Inputs I: mean(mu) and sd(sigma) parameter of process
    #Inputs II: Time Interval [0,T], Steps in interval n
    #Inputs III: Number of Simulated BMs - nsim
    #Output: Geometric Brownian Motion
    time <- seq(0,T,by=(T/n))
    B <- t(sim.BM(T, n, nsim))
    GBM <- matrix(0,nsim, dim(B)[2])
    for (i in 1:nsim){
        GBM[i,] <- exp(sigma*B[i,] + mu*time)
    }
        GBM <- cbind(time,t(GBM))
        data.frame(GBM)
}

######################################################
#Simulation of Brownian Motion with Drift
######################################################

sim.BM.drift <- function(mu, sigma, T, n, nsim=1){
    #Inputs I: mean(mu) and sd(sigma) parameter of process
    #Inputs II: Time Interval [0,T], Steps in interval n
    #Inputs III: Number of Simulated BMs - nsim
    #Output: Brownian Motion with Drift
    time <- seq(0,T,by=(T/n))
    B <- t(sim.BM(T, n, nsim))
    BM.drift <- matrix(0,nsim, dim(B)[2])
    for (i in 1:nsim){
        BM.drift[i,] <- sigma*B[i+1,] + mu*time
    }
        BM.drift <- cbind(time,t(BM.drift))
        data.frame(BM.drift)

}

######################################################
#Simulation of Brownian Bridge
######################################################
sim.BBridge <- function(T, n, nsim=1){
    #Inputs I: Time Interval [0,T], Steps in interval n
    #Inputs II: Number of Simulated BMs - nsim
    #Output: Brownian Bridge
    time <- seq(0,T,by=(T/n))
    B <- t(sim.BM(T, n, nsim))[-1,]
    B <- matrix(B,nrow=nsim)
    time.mat <- t(replicate(nsim, time))
    BBridge <- B  - B[,2]*time.mat
    BBridge <- cbind(time,t(BBridge))
    data.frame(BBridge)
}

######################################################
#Simulation of Squared Brownian Motion - Time
######################################################
sim.sqBM <- function(T, n, nsim=1){
    #Inputs I: Time Interval [0,T], Steps in interval n
    #Inputs II: Number of Simulated BMs - nsim
    #Output: Squared BM minus t
    time <- seq(0,T,by=(T/n))
    B <- t(sim.BM(T, n, nsim))
    sqBM <- B[-1,]^2  - t(replicate(nsim, time))
    sqBM <- cbind(time,t(sqBM))
    data.frame(sqBM)
}

######################################################
#Function for Plotting
######################################################
plot.BM <- function(BM){
    #Input: Realizations of Stochastic Process(es)
    #Output: Nice plot
    newdf<-melt(BM,'time')
    plot <- ggplot(newdf,aes(x=time,y=value,group=variable,color=variable)) +
                geom_line() +
                theme(legend.position="none") +
                ylab("Stochastic Process") +
                xlab("Time") +
                theme(plot.title = element_text(hjust = 0.5, lineheight=.8))
    plot
}
```

## Brownian Motion

```r
######################################################
#PLOTTING - Figure 1
######################################################
BM_5_50_1 <- plot.BM(sim.BM(T=5, n=50, nsim=1)) +
    ggtitle(TeX('Single Brownian Motion - T=5, n=50'))
BM_5_100_1 <- plot.BM(sim.BM(T=5, n=100, nsim=1)) +
    ggtitle(TeX('Single Brownian Motion - T=5, n=100'))
BM_5_1000_1 <- plot.BM(sim.BM(T=5, n=1000, nsim=1)) +
    ggtitle(TeX('Single Brownian Motion - T=5, n=1000'))
BM_5_50_50 <- plot.BM(sim.BM(T=5, n=50, nsim=50)) +
    ggtitle(TeX('50 Brownian Motions - T=5, n=50'))
BM_5_100_50 <- plot.BM(sim.BM(T=5, n=100, nsim=50)) +
    ggtitle(TeX('50 Brownian Motions - T=5, n=100'))
BM_5_1000_50 <- plot.BM(sim.BM(T=5, n=1000, nsim=50)) +
    ggtitle(TeX('50 Brownian Motions - T=5, n=1000'))

multiplot(BM_5_50_1, BM_5_100_1, BM_5_1000_1, BM_5_50_50, BM_5_100_50, BM_5_1000_50, cols=2)
```

By observing these graphs we note that we start by unbiased random walk movement in continuous time  but as we  increase N, i.e. as n, the motion appears more and more volatile, and the process converges to Brownian motion: $S_n(t), t\epsilon(0,T)\rightarrow B_t, t\epsilon(0,T)$.

## Geometric Brownian Motion

```r
######################################################
#PLOTTING - Figure 2
######################################################

GBM_1_1 <- plot.BM(sim.GBM(mu=-0.5,sigma=1,T=1,n=1000,nsim=1)) +
    ggtitle(TeX('GBM: $X_t = e^{\\sigma B_t + \\mu t}$ with $\\mu = -0.5$, $\\sigma = 1$'))
GBM_2_1 <- plot.BM(sim.GBM(mu=0.5,sigma=1,T=1,n=1000,nsim=1)) +
    ggtitle(TeX('GBM: $X_t = e^{\\sigma B_t + \\mu t}$ with $\\mu = 0.5$, $\\sigma = 1$'))
BMD_1_1 <- plot.BM(sim.BM.drift(mu=1,sigma=0.1,T=1,n=1000,nsim=1)) +
    ggtitle(TeX('BMDrift: $X_t = \\sigma B_t + \\mu t$ with $\\mu = 1$, $\\sigma = 0.1$'))
BMD_2_1 <- plot.BM(sim.BM.drift(mu=0.1,sigma=1,T=1,n=1000,nsim=1)) +
    ggtitle(TeX('BMDrift: $X_t = \\sigma B_t + \\mu t$ with $\\mu = 0.1$, $\\sigma = 1$'))
BBridge_1 <- plot.BM(sim.BBridge(T=1, n=1000, nsim=1)) +
    ggtitle(TeX('Brownian Bridge: $X_t = B_t - tB_1$'))
sqBM_1 <- plot.BM(sim.sqBM(T=1, n=1000, nsim=1)) +
    ggtitle(TeX('$X_t = B_t^2 - t$'))

multiplot(GBM_1_1, BMD_1_1, BBridge_1, GBM_2_1, BMD_2_1, sqBM_1, cols=2)
```
